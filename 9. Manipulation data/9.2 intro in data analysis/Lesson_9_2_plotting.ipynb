# Pandas plotting

The most popular library in Python to build plots is `matplotlib`, some other libraries are built on top of it, including `pandas`. That means that `pandas` offers a more user-friendly interface to matplotlib functions.

You need to `pip install matplotlib` manually in your own environment, before you can access graph features.

Let's go through some basic types of plots. We will use "Tips" dataset ([data](https://raw.githubusercontent.com/mwaskom/seaborn-data/master/tips.csv), [source](https://rdrr.io/cran/reshape2/man/tips.html)) and "Dowjones" ([data](https://raw.githubusercontent.com/mwaskom/seaborn-data/master/dowjones.csv), [source](https://fred.stlouisfed.org/series/M1109BUSM293NNBR)).

Pandas uses the `plot()` method to create diagrams. It accepts a lot of parameters, refer to [docs](https://pandas.pydata.org/docs/reference/api/pandas.DataFrame.plot.html).

```python
import pandas as pd

df_tips = pd.read_csv('https://raw.githubusercontent.com/mwaskom/seaborn-data/master/tips.csv')
df_dj = pd.read_csv('https://raw.githubusercontent.com/mwaskom/seaborn-data/master/dowjones.csv', parse_dates=['Date'])
```

## Line plot

This is the default plot kind.

```python
df_dj.plot(x='Date', y='Price') # same as df_dj.plot(x='Date', kind='line')
```

```python
df_tips.head()
```

## Bar

To compare categorical or descrete data, we can use bar plot.

```python
df_tips.groupby('time').agg({'total_bill': 'mean'}).plot(kind='bar')
```

```python
df_tips.groupby('time').agg({'total_bill': 'mean'}).plot(kind='barh', title='Mean bill by time')
```

## Histogram

Histograms are a useful tool to inspect data distribution in the dataset.

Histograms have also their own function [hist()](https://pandas.pydata.org/docs/reference/api/pandas.DataFrame.hist.html).

```python
df_tips.plot(kind='hist', y='total_bill')
```

## Box

Bot plot displays the distribution of a dataset through five key statistical measures: the minimum, first quartile (Q1), median, third quartile (Q3), and maximum.

It is more convenient to use [boxplot()](https://pandas.pydata.org/docs/reference/api/pandas.DataFrame.boxplot.html) function.

```python
df_tips.boxplot('total_bill', by='time')
```

## Pie

Pie plot divides data into slices, each representing a proportion or percentage of the whole.

```python
df_tips.groupby('time').agg({'total_bill': 'sum'}).plot(kind='pie', y='total_bill')
```
